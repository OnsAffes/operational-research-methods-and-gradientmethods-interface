### Implémentation de Simplexe Phase I

```python
class LinearModel: 
    
    def __init__(self, A = np.empty([0,0]), b = np.empty([0,0]), c = np.empty([0,0]), minmax = "MAX"): #Initialisation des attributs de la classe
        self.A = A
        self.b = b
        self.c = c
        self.x = [float(0)] * len(c)
        self.minmax = minmax
        self.printIter = True
        self.optimalValue = None
        self.transform = False
        self.feasible = True
        self.bounded = True
    
    def addA(self, A): # Initialisation de la matrice A
        self.A = A
        
    def addB(self, b): # Initialisation de la matrice B
        self.b = b
        
    def addC(self, c): # Initialisation de la matrice C
        self.c = c
        self.transform = False
    
    def setObj(self, minmax): # Modification de la varible minmax que avec MIN ou MAX
        if(minmax == "MIN" or minmax == "MAX"):
            self.minmax = minmax
        else:
            print("Invalid objective.") # Sinon afficahge d'un message d'erreur
        self.transform = False
            
    def setPrintIter(self, printIter): # Modification de la varible printIter 
        self.printIter = printIter
            
    def printSoln(self):
        if(self.feasible): # Tester la faisablité de la solution 
            if(self.bounded):
                print("Coefficients: ")
                print(self.x) # Affichage des coefficients
                print("Valeur Optimale: ")
                print(self.optimalValue) # Affichage de la valeur optimale
            else:
                print("Problème non bounded") # Affichage d'un message d'erreur
        else:
            print("Problème non faisable") # Affichage d'un message d'erreur
        
    def printTableau(self, tableau):
        
        print("ind \t\t", end = "")
        for j in range(0, len(self.c)):
            print("x_" + str(j), end = "\t")
        for j in range(0, (len(tableau[0]) - len(self.c) - 2)):
            print("s_" + str(j), end = "\t")
        
        print()
        for j in range(0, len(tableau)):
            for i in range(0, len(tableau[0])):
                if(not np.isnan(tableau[j, i])):
                    if(i == 0):
                        print(int(tableau[j, i]), end = "\t")
                    else:
                        print(round(tableau[j, i], 2), end = "\t")
                else:
                    print(end = "\t")
            print()
            
    def getTableauPhase1(self): # Méthode qui utilise le tableau de la phase I
        
        # Construire le tableau de départ
        
        numVar = len(self.c)
        numArtificial = len(self.A)
        
        t1 = np.hstack(([None], [0], [0] * numVar, [0] * numArtificial)) # Ajout de la variable artificielle
                    
        basis = np.array([0] * numArtificial)
        
        for i in range(0, len(basis)):
            basis[i] = numVar + i
        
        A = self.A
        
        if(not ((numArtificial + numVar) == len(self.A[0]))):
            B = np.identity(numArtificial)
            A = np.hstack((self.A, B))
        
        t2 = np.hstack((np.transpose([basis]), np.transpose([self.b]), A)) # Faire le transposé pour t2
        
        tableau = np.vstack((t1, t2)) 
        
        for i in range(1, len(tableau[0]) - numArtificial): #Remplissage de tableau
            for j in range(1, len(tableau)):
                if(self.minmax == "MIN"):
                    tableau[0, i] -= tableau[j, i]
                else:
                    tableau[0, i] += tableau[j, i]
        
        tableau = np.array(tableau, dtype ='float')
        
        return tableau
    
    def driveOutAV(self, tableau): #
        avPresent = False
        avInd = []
        
        #supprimer les variables d'écart
        tableau = np.delete(tableau, obj = np.s_[2 + len(self.c):], axis = 1)
        
        #suppression de lignes redondantes
        for i in range(1, len(tableau)):
            if tableau[i, 0] > len(self.c) - 1:
                
                redundant = True
                for j in range(2, len(self.c) + 2):
                    if (tableau[i, j] != 0):
                        redundant = False
                        avInd.append([i, False])
                        break
                avInd.append([i, redundant])
                
        if(len(avInd) == 0):  # Si la longueur de avInd égale à 0, on affiche ce message
            print("\nIl n y pas de variable artificielle qui va sortir\n")
            return tableau
        
        for key in avInd:
            index, redundant = key
            if(redundant):
                tableau = np.delete(tableau, obj = index, axis = 0)
                avInd.remove(key)
            else:
                n = -1
                for i in range(2, len(tableau[0]) - 1):
                    if (tableau[r, i] != 0):
                        n = i
                        break
                tableau = pivot(tableau, index, n)
        
        return tableau
    
    def getTableauPhase2(self, tableau): # Méthode qui utilise le tableau de la phase I
        
        for i in range(0, len(self.c) - 1):
            tableau[0, i + 2] = self.c[i]
            
        tableau[0, 1] = 0
        
        a = np.empty(0)
        for value in tableau[0, 1:]:
            a = np.append(a, value)
        
        for j in range(1, len(tableau)):
            for i in range(1, len(tableau[0])):
                if(self.minmax == "MIN"):
                    a[i - 1] -= tableau[0, int(tableau[j, 0] + 2)] * tableau[j, i]
                else:
                    a[i - 1] += tableau[0, int(tableau[j, 0] + 2)] * tableau[j, i]
            
        tableau[0, 1:] = a
            
        return tableau
    
    def pivot(self, tableau, r, n):
        
        pivot = tableau[r, n]
        
        # effectuer des opérations sur les lignes
        # diviser la rangée de pivot avec l'élément de pivot
        tableau[r, 1:] = tableau[r, 1:] / pivot 
            
        # faire pivoter d'autres lignes
        for i in range(0, len(tableau)): 
            if i != r:
                mult = tableau[i, n] / tableau[r, n]
                tableau[i, 1:] = tableau[i, 1:] - mult * tableau[r, 1:]
                
        return tableau
        
              
    def optimize(self):
        
        tableau = self.getTableauPhase1()
        
        print("Phase 1:\n")
        
        tableau = self.simplex(tableau)
        
        if(not self.bounded):
            return
        
        if(tableau[0, 1] != 0):  # Tester la faisablité de la solution
            self.feasible = False # La variable de test de faisabilité reçoit False
            print("Problème irréalisable ; Pas de solution") # Affichage de message d'erreur
            return
                
        tableau = self.driveOutAV(tableau)
        
        tableau = self.getTableauPhase2(tableau)
        
        tableau = self.simplex(tableau)
        
        if(not self.bounded):
            return
       
        self.x = np.array([0] * len(self.c), dtype = float)
        
        # Enregistrer les coefficients
        for key in range(1, (len(tableau))):
            if(tableau[key, 0] < len(self.c)):
                self.x[int(tableau[key, 0])] = tableau[key, 1]
                
        self.optimalValue = -1 * tableau[0,1]

        
    def simplex(self,tableau):
        global verif,P1
        if(self.printIter == True):
            print("Starting Tableau:")
            self.printTableau(tableau)
        
        # Supposer que la base initiale n'est pas optimale, que le problème est réalisable et que le problème est borné
        optimal = False
        feasible = True
        bounded = True

        # Garder une trace des itérations pour l'affichage
        iter = 1

        while(True):
            
            if(self.printIter == True):
                print("----------------------------------")
                print("Iteration :", iter)
                self.printTableau(tableau)
                
            if(self.minmax == "MAX"):
                for profit in tableau[0, 2:]:
                    if profit > 0:
                        optimal = False
                        break
                    optimal = True
            else:
                for cost in tableau[0, 2:]:
                    if cost < 0:
                        optimal = False
                        break
                    optimal = True

            # si toutes les directions entraînent une diminution des bénéfices ou une augmentation des coûts
            if optimal == True: 
                 break
            
            # n variable entre la base, compte pour l'indexation du tableau
            if (self.minmax == "MAX"):
                n = tableau[0, 2:].tolist().index(np.amax(tableau[0, 2:])) + 2
            else:
                n = tableau[0, 2:].tolist().index(np.amin(tableau[0, 2:])) + 2

            # test du rapport minimum, r base de feuilles variables
            minimum = 99999
            r = -1

            for i in range(1, len(tableau)): 
                if(tableau[i, n] > 0):
                    val = tableau[i, 1]/tableau[i, n]
                    if val<minimum: 
                        minimum = val 
                        r = i
                            
            pivotElement = tableau[r, n] 
            
            if(self.printIter):
                print("Pivot Column:", n)
                print("Pivot Row:", r)
                print("Pivot Element: ", pivotElement)
                print("New Basis: " + str(tableau[r, 0]))

            for element in tableau[1:, n]:
                if (element != 0):
                    self.bounded = True
            
            if (not self.bounded):
                print("Unbounded; No solution.")
                return
            tableau = self.pivot(tableau, r, n)


            # Nouvelle variable de base
            tableau[r, 0] = n - 2
            
            
            iter += 1
            
        
        if(self.printIter == True):
            print("----------------------------------")
            print("Final Tableau reached in", iter, "iterations")
            self.printTableau(tableau) # Affichage de dernier tableau et le nombre d'itérations
        else:
            print("Solved")
        
        verif=1
        P1=tableau.copy()
        return tableau
  
 
```

```python
def phase1(prob,C, Sol,V,Mode,Ci,con) : 
    model1 = LinearModel() # Creation d'une instance de la classe LinearModel
    c=[]
    # Remplissage de la matrice c avec ZMAX et la base canonique (0 0 0) 
    c =  c + [- prob[0][x] for x in range(0,V)] 
    for i in range (C) :
        c.append(0)
    ln=0;
    col=0;
    # Remplissage de la matrice A avec les coefficients des x dans les contraintes et une base canonique pour chaque ligne
    # tout depend des contraintes >= / <= / =
    A= np.zeros((C,(C+V)), dtype=float)
    X=[]
    for i in range (1,C+1) :
        X =  X + [[prob[i][x] for x in range(V)]]  
        
    #Copier les valeur X au A
    while(ln<C):
        while(col<V):
            A[ln][col]=X[ln][col]
            col=col+1
        col=0;
        ln=ln+1;

    # Valeur du base canonique
    # Ajouter la base canonique
    cont = 0
    while(cont < C):
        if(con[cont] == 3):
            A[cont][cont+V]= -1
        elif(con[cont] == 2):
            A[cont][cont+V]= 0
        else:
            A[cont][cont+V]= 1
        cont=cont+1 
    
    b=Ci
    b=b[1:len(b):1]
    
    model1.addA(A) # Initialisation de la matrice A
    model1.addB(b) # Initialisation de la matrice b
    model1.addC(c) # Initialisation de la matrice c
    model1.setObj("MIN")
    model1.setPrintIter(True)

    
    model1.optimize()
    print("\n ")
    model1.printSoln()
```
